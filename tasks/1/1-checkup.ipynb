```python
def raw_polynomials(x):
    P1 = -9804 * x**32 + 4423.3 * x**16 - 8.7654 * x**8
    P2 = 9.09 * x**9 - 9.09 * x**3 + 9.09 * x

    return P1, P2, P1 * P2
```

```python
def step_factorization(x):
    P1 = [-9804 * x**16 + 4423.3]
    P1.append(P1[0] * x**8 - 8.7654)
    P1.append(P1[1] * x**8)

    P2 = [x**6 - 1]
    P2.append(P2[0] * x**2 + 1)
    P2.append(P2[1] * 9.09 * x)

    return P1, P2, P1[-1] * P2[-1]
```

```python
num = 5.252

print(*raw_polynomials(num), sep="\n")
print()
print(*step_factorization(num), sep="\n")
```
